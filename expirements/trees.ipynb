#### Удаление корреляций

```python
X = train_df.drop(['search_id','target'],axis =1)
group_id = train_df['search_id']
y = train_df['target']
```

```python
corr_f = X.corr()
upper = corr_f.where(np.triu(np.ones(corr_f.shape), k=1).astype(bool))
upper
```

```python
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
len(to_drop)
```

```python
corr_features = X[to_drop]
```

```python
X_1 = X.drop(to_drop,axis = 1)
X_1.shape
```

#### Сплит

```python
X_train, X_val, y_train, y_val = train_test_split(X_1, y, test_size=0.3, random_state=42)
```

#### Нормализация

```python
from sklearn.preprocessing import StandardScaler
```

```python
normalizer = StandardScaler()
X_train_norm = normalizer.fit_transform(X_train)
X_val_norm = normalizer.transform(X_val)
```

#### Sampling

minority/all = 0.2 => minoity/majority = 0.25

```python
from imblearn.over_sampling import SMOTE, ADASYN
sm = SMOTE(random_state=42,sampling_strategy=0.07)
# ad = ADASYN(random_state=42)
X_res, y_res = sm.fit_resample(X_train_norm,y_train)
y_res.value_counts(normalize=True)
```

```python
X_res = pd.DataFrame(X_res,columns=X_train.columns)
X_val_norm = pd.DataFrame(X_val_norm,columns=X_val.columns)
```

### Randomforest classifier

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import ndcg_score
```

```python
rfc = RandomForestClassifier(random_state=42,n_jobs = -1,n_estimators=150,class_weight='balanced')
```

```python
param_grid_rfc = {
        'max_depth': [10,20,30],
        'min_samples_leaf': [1,2,3],
        'criterion' : ['entropy','gini'],
}
```

```python
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
grid_search_rfc = GridSearchCV(rfc, param_grid_rfc, cv=cv_strategy, scoring='f1',verbose=2)
grid_search_rfc.fit(X_res,y_res)
```

```python
grid_search_rfc.best_params_
```

```python
best_model_rfc = grid_search_rfc.best_estimator_
preds_rfc = best_model_rfc.predict(X_val_norm)
```

```python
pd.DataFrame(preds_rfc,columns=['preds']).value_counts(normalize=True)
```

```python
y_val.value_counts(normalize=True)
```

```python
y_res.value_counts(normalize=True)
```

```python
print('Тестовая точность: {}'.format(accuracy_score(y_val,preds_rfc)))
print('Тестовая roc-auc: {}'.format(roc_auc_score(y_val,preds_rfc)))
print('Тестовый f1: {}'.format(f1_score(y_val,preds_rfc)))
print('Тестовый NDCG: {}'.format(ndcg_score([y_val],[preds_rfc])))
```
